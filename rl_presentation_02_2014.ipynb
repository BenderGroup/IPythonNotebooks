Why Python?
---

- Python is used everywhere, and as such has needed to build up libraries to interact with code written in many other languages.  Therefore, it is probably the easiest tool to interact with other languages. 
- Easily readable - has been called executable pseudocode.  This means even non Pythonistas may be able to follow the code 
- Extremely *terse* (python code is shorter than the equivalent in almost all languages).  This allows for a large boost to productivity.
- As it is built on C/C++, properly used it can be very fast
- It is on the way to becoming the *de facto* language of data science, starting to beat out Matlab, and challenging R.  With this is coming well supported, modern libraries 
- It is completely free and open source (unlike Matlab, .NET, etc.)
- It is cross platform, is used on anything from phones to supercomputers.
- The IPython project!

## IPython

- IPython at its core is an advanced Python interpreter.
- It is designed to be used to explore data, rather than writing long scripts.

- Part of the Scientific Python libraries.  The other parts are:
    - NumPy (provides fast c array object)
    - SciPy (implements many scientifically relevent functions such as FFT)
    - Matplotlib (high quality graphing library)
    - Sympy (symbolic mathematics)
    - Pandas (provides Python with a R DataFrame object)

- Features:
    - Automatic caching of input and output
    - Easier access to help and documentation
    - Very thorough tab completion
    - Extensions named magic functions provide easy access to useful functionality, and can be extended 
    - Access to other languages
    - Running on a remote server
    - Simple parallel processing support

Running IPython
------

- IPython can be run from the command line simply using `ipython`
- It can also be run in a client-server model.
- A server can be run from the command line, or on a remote server, such as a cluster.
- Allows for access to your machine from anywhere where you have a reasonably up to date browser, much better than ssh!

- There are many clients, including:
    - the traditional command line tool
    - a qt based console
    - *IPython notebook*

External Language Support
------

Python as a language is regularly used as 'glue code', calling code from many different languages.  This is usually implemented via modules in regular Python, and is clunky to use.  IPython makes this incredibly easy, by wrapping the language in magic functions.

For example, in R:

```python
%load_ext rmagic
```

```python
%%R 
X <- 1:25; 
print(X); 
cat("Mean of X:", mean(X));
```

It is possible to pass in Python variables, and to retrieve them out. 

```python
#python code
python_list = numpy.random.randn(25); python_list
```

```python
%%R -i python_list -o R_output #R code imports the python data, and can process it

print(head(python_list))
R_output <- mean(python_list)
print(lm(python_list~X))
```

```python
Image(url='http://3.bp.blogspot.com/-4laOSUp4TzQ/Uj0t7ZZq6DI/AAAAAAAABFY/1GpO_uJoRWg/s1600/logo.lrg.png', )
```

- RDKit is a newer cheminformatics library which is geared specifically towards cheminformatics.  
- It supports a variety of fingerprints and descriptors, and provides python bindings.
- There is also a PostgreSQL plugin, that supports server-side similarity searches, fingerprint calculations etc.  

- Scikit learn is a machine learning library.  It provides many different models, and works well with RDKit fingerprints.

```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import PandasTools
```

We can execute SQL queries directly in the notebook. Load up sql tools, and connect to a postgresql database:

```python
%load_ext sql
%sql postgresql://RichLewis@localhost/mix_model
```

This database was produced from ChEMBL, but stripped of any excess data and optimized to make it as fast as possible.  It currently contains 3 tables from ChEMBL (compound, sar, target) and 2 from rdkit, along with cached morgan fingerprints. This is shown using a cell level (%%) sql magic function:

```python
%%sql 
SELECT 
    table_schema,table_name
FROM 
    information_schema.tables 
WHERE 
    table_schema IN ('public', 'rdk') 
ORDER BY 
    table_schema,table_name
```

It is in as neutral a format as possible to allow for extra targets, compounds and SARs to be added.

In this example, we select all molecules that we are confident bind to prothrombin (target_id 95), along with their binding affinity (pValue)

```python
prothrombin_sar = %sql SELECT external_id, relation, activity_type, assay_type, "pValue", mols.m \
FROM \
    sar NATURAL JOIN rdk.mols NATURAL JOIN compound \
WHERE target_id = 95 \
    AND confidence_score = 9
prothrombin_sar[:10]
```

This is not particularly helpful to visualize molecules, so we can use RDKit to draw the molecules into a table.  To do this, we can use a pandas dataframe:

```python
import pandas
from rdkit.Chem import PandasTools

df = prothrombin_sar.DataFrame()
df = df.set_index('external_id')
PandasTools.AddMoleculeColumnToFrame(df,  smilesCol='m', molCol='Molecules')
df = df.drop('m', 1)
df.head()
```

We can plot the pValue (the pKi, pIC50 etc. values combined) distribution in a histogram using matplotlib:

```python
array = numpy.array(df['pValue'], dtype=float)
num_bins = 15

# the histogram of the data
n, bins, patches = plt.hist(array, num_bins, facecolor='green', alpha=0.5)
plt.xlabel('pValue')
plt.ylabel('Density')
plt.title(r'Plot of bioactivities for {target}'.format(target = 'Prothrombin'))
plt.show()
```

Creating the target prediction model
-----
We now need to set up a feature vector and corresponding class labels. In this case the problem can be reduced to a two class (Binder or non-binder) problem.  We shall use morgan fingerprints as features.  These have been precalculated, so we start by retrieving them from database:

```python
result = %sql \
SELECT \
compound_id, "pValue", fingerprint \
FROM sar NATURAL JOIN "morgan(2,2048)" \
WHERE target_id = 95
```

Insert them into arrays:

```python
import cPickle
X = np.zeros((len(result),2048),dtype=uint8)
y = np.zeros((len(result)),dtype=float)
for count, line in enumerate(result):
    compound_id, pValue, pickle = line
    fp = cPickle.loads(str(pickle))
    X[:][count] = fp
    y[count] = pValue
print 'Shape of X:', X.shape #dimensions
print 'Shape of y:', y.shape
```

We binarize into two classes, by 6.5 to achieve approximately equal class sizes:

```python
y_binarized = y > 6.5 #binarize 
y_binarized
```

We can now use one of scikit-learn's classification algorithms in a 5 fold cross validation, plot a confusion matrix and summary of the model produced.  

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn import cross_validation
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

classifier = BernoulliNB() 

cv = cross_validation.StratifiedKFold(y_binarized, n_folds = 5)

pred = numpy.zeros(len(y_binarized),dtype=bool)

for train, test in cv:
    pred[test] = classifier.fit(X[train], y_binarized[train]).predict(X[test])

cm = confusion_matrix(y_binarized, pred)
print 'Confusion matrix'
print cm

pl.matshow(cm)
pl.title('Confusion matrix')
pl.colorbar()
pl.ylabel('True label')
pl.xlabel('Predicted label')
pl.show()
print classification_report(y_binarized, pred)
```

We can also generate ROC curves:

```python
from sklearn.metrics import roc_curve, auc

pred_prob = numpy.zeros((len(y_binarized),2),dtype=float)

for train, test in cv:
    pred_prob[:][test] = classifier.fit(X[train], y_binarized[train]).predict_proba(X[test])

fpr, tpr, thresholds = roc_curve(y_binarized, pred_prob[:, 1])
roc_auc = auc(fpr, tpr)
pl.plot(fpr, tpr, lw=1, label='ROC (area = %0.2f)' % (roc_auc))
pl.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')
pl.xlim([-0.05, 1.05])
pl.ylim([-0.05, 1.05])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('ROC Curve')
pl.legend(loc="lower right")
pl.show()
```

We can use different classifiers:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

classifiers = {'lr':LogisticRegression(), 
               'rf':RandomForestClassifier(), 
               'dt':DecisionTreeClassifier(),
               'knn':KNeighborsClassifier(),
               'bnb':BernoulliNB()}

for name, classifier in classifiers.iteritems():
    
    pred_prob = numpy.zeros((len(y_binarized),2),dtype=float)

    for train, test in cv:
        pred_prob[:][test] = classifier.fit(X[train], y_binarized[train]).predict_proba(X[test])
    
    fpr, tpr, thresholds = roc_curve(y_binarized, pred_prob[:, 1])
    roc_auc = auc(fpr, tpr)
    pl.plot(fpr, tpr, lw=1, label='ROC %s (area = %0.2f)' % (name, roc_auc))
    
pl.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck',)
pl.xlim([-0.05, 1.05])
pl.ylim([-0.05, 1.05])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('ROC Curve')
pl.legend(loc="lower right")
pl.show()
    
```

For this target class at least, Naive Bayes doesn't appear to be a particularly good classifier.  Perhaps logistic regression or Random Forests would be better.

Predicting binding orientation with Target Prediction
-----
One of the ligands, compound_id = 19143, is Benzothiophene Inhibitor 4.  There are crystal structures of it cocrystallized with Thrombin.  We can build a model specifically excluding it, and predict whether it binds:


```python
result = %sql \
SELECT \
compound_id, "pValue", fingerprint \
FROM sar NATURAL JOIN "morgan(2,2048)" \
WHERE target_id = 95 \
AND compound_id <> 19143
```

```python
X = np.zeros((len(result),2048),dtype=uint8)
y = np.zeros((len(result)),dtype=float)
for count, line in enumerate(result):
    compound_id, pValue, pickle = line
    fp = cPickle.loads(str(pickle))
    X[:][count] = fp
    y[count] = pValue

y_binarized = y > 6.5
```

We can calculate the fingerprints in python using the rdkit library:

```python
binder = Chem.MolFromSmiles('Oc1ccc2c(Cc3ccc(CN4CCCC4)c(Br)c3)c(sc2c1)-c1ccc(OCCN2CCCC2)cc1'); binder
```

```python
fp = Chem.AllChem.GetMorganFingerprintAsBitVect(binder,2,nBits=2048)
fp.ToBitString()
```

Using a random forest classifier, our model can predict if it binds. Fortunately it is predicted active!

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X,y_binarized)
print classifier.predict(fp)
```

We can then use the rdkit library to visualise the contributions of the different fingerprints to its probability of binding.

```python
from rdkit.Chem.Draw import SimilarityMaps
def getProba(fp, predictionFunction):
  return predictionFunction(fp)[0][1]
fig, maxweight = SimilarityMaps.GetSimilarityMapForModel(binder, SimilarityMaps.GetMorganFingerprint, lambda x: getProba(x, classifier.predict_proba))
```

It appears that a lot of the molecule contributes.  Notably, the O linker and the Br do not.  Looking at the crystal structure, these appear to be in least contact with the protein.
